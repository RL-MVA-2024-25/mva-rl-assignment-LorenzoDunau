```python
import numpy as np
from scipy.optimize import minimize

# Définir la fonction à minimiser
def f(t):
    t1, t2 = t  # t est un tableau de deux éléments
    return 0.3 * (t1 + t2) / (t1 + t2 + 100) - 0.25 * (t1 + t2) / (t1 + t2 + 500)

# Valeurs initiales pour t1 et t2
initial_guess = [1, 1]

# Contraintes pour que t1 et t2 soient positifs
bounds = [(1, None), (1, None)]  # (min, max) pour chaque variable; None signifie sans limite supérieure

# Utilisation de scipy.optimize.minimize avec la méthode L-BFGS-B pour inclure des bornes
result = minimize(f, initial_guess, method='L-BFGS-B', bounds=bounds)

# Afficher les résultats
print("Minimum trouvé pour t1 et t2 : ", result.x)
print("Valeur minimale de la fonction : ", result.fun)
```
